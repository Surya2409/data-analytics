## Importing Libraries

```python
from bs4 import BeautifulSoup
```

## Scraping Website

```python
url="https://www.flipkart.com/search?q=apple%20macbook%20air%20m1&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

## Response

```python
response=requests.get(url)
```

```python
response
```

## Parsing

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
names=soup.select('._4rR01T')
```

```python
names
```

```python
laptop_names=[]
```

```python
for item in names:
    laptop_names.append(item.text)
    laptop_names
```

```python
laptop_names
```

```python
price=soup.select('._1_WHN1')
```

```python
price
```

```python
laptop_price=[]
```

```python
for item in price:
    laptop_price.append(item.text)
    laptop_price
```

```python
 laptop_price
```

```python
ram=soup.select('.rgWa7D:nth-child(2)')
```

```python
ram
```

```python
laptop_ram=[]
```

```python
for item in ram:
    laptop_ram.append(item.text)
    laptop_ram
```

```python
laptop_ram
```
